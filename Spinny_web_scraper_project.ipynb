```python
#connect to websites
url="https://www.spinny.com/blog/index.php/top-selling-car-brands-in-india/"
```

```python
page= requests.get(url)
soup= BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
soup.find_all('tr',class_="wptb-row")[1]
```

```python
tabletitles= soup.find_all('tr',class_="wptb-row")[1]
tabletitles
```

```python
table_titles=[title.text.strip() for title in tabletitles]
table_titles
```

```python
df= pd.DataFrame(columns=table_titles)
df
```

```python
column_data=soup.find_all('tr')
column_data
```

```python
for row in column_data[1:]:
  row_data= row.find_all('td')
  individual_row_data=[data.text.strip() for data in row_data]

  length=len(df)
  df.loc[length]=individual_row_data

```
